```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import layers, losses
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
import time

def noise(array):
    """
    Adds random noise to each image in the supplied array.
    """

    noise_factor = 0.1
    noisy_array = array + noise_factor * np.random.normal(
        loc=0.0, scale=10.0, size=array.shape
    )

    return np.clip(noisy_array, 0.0, 1.0)

def display(array1, array2, array3):
    """
    Displays ten random images from each one of the supplied arrays.
    """
    n = 10
    psnr = getpsnr(array1, array3).numpy()
    indices = np.random.randint(len(array1), size=n)
    images1 = array1[indices, :]
    images2 = array2[indices, :]
    images3 = array3[indices, :]
    psnr = psnr[indices]
    plt.figure(figsize=(20, 6))
    for i, (image1, image2, image3) in enumerate(zip(images1, images2, images3)):
        ax = plt.subplot(3, n, i + 1)
        plt.imshow(image1.reshape(32, 32, 3))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        ax = plt.subplot(3, n, i + 1 + n)
        plt.imshow(image2.reshape(32, 32, 3))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        ax = plt.subplot(3, n, i + 1+ (2 *n))
        plt.imshow(image3.reshape(32, 32, 3))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        print(psnr[i])
    plt.show()
    return

def getpsnr(array1, array2):
    psnr = tf.image.psnr(array1, array2, max_val=255)
    return psnr


(x_train, _), (x_test, _) = tf.keras.datasets.cifar10.load_data()

x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

print(x_train.shape)
print(x_test.shape)

noizy_x_train = noise(x_train)
noizy_x_test = noise(x_test)

input = layers.Input(shape=(32, 32, 3))

x = layers.Conv2D(32, (3, 3), activation="relu", input_shape=(32, 32, 3), padding='same')(input)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
x = layers.Conv2D(32, (3, 3), activation="relu", padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding="same")(x)

x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2D(3, (3, 3), activation="sigmoid", padding="same")(x)

model = Model(input, x)
model.compile(optimizer="adam", loss=losses.MeanSquaredError(), metrics=['accuracy'])
model.fit(noizy_x_train, x_train, epochs=10, batch_size=64, validation_data=(noizy_x_test, x_test))
start = time.time()
predictions = model.predict(noizy_x_test)
print("Time Taken : %f" % (time.time() - start))
display(x_test, noizy_x_test, predictions)


x = layers.Conv2D(64, (3, 3), activation="relu", input_shape=(32, 32, 3), padding='same')(input)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
x = layers.Conv2D(64, (3, 3), activation="relu", padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding="same")(x)

x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2D(3, (3, 3), activation="sigmoid", padding="same")(x)

model = Model(input, x)
model.compile(optimizer="adam", loss=losses.MeanSquaredError(), metrics=['accuracy'])
model.fit(noizy_x_train, x_train, epochs=10, batch_size=64, validation_data=(noizy_x_test, x_test))
start = time.time()
predictions = model.predict(noizy_x_test)
print("Time Taken : %f" % (time.time() - start))
display(x_test, noizy_x_test, predictions)


x = layers.Conv2D(32, (3, 3), activation="relu", input_shape=(32, 32, 3), padding='same')(input)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
x = layers.Conv2D(32, (3, 3), activation="relu", padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
x = layers.Conv2D(32, (3, 3), activation="relu", padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding="same")(x)

x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2D(3, (3, 3), activation="sigmoid", padding="same")(x)

model = Model(input, x)
model.compile(optimizer="adam", loss=losses.MeanSquaredError(), metrics=['accuracy'])
model.fit(noizy_x_train, x_train, epochs=10, batch_size=64, validation_data=(noizy_x_test, x_test))
start = time.time()
predictions = model.predict(noizy_x_test)
print("Time Taken : %f" % (time.time() - start))
display(x_test, noizy_x_test, predictions)

x = layers.Conv2D(64, (3, 3), activation="relu", input_shape=(32, 32, 3), padding='same')(input)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
x = layers.Conv2D(64, (3, 3), activation="relu", padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
x = layers.Conv2D(64, (3, 3), activation="relu", padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding="same")(x)

x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation="relu", padding='same')(x)
x = layers.Conv2D(3, (3, 3), activation="sigmoid", padding="same")(x)

model = Model(input, x)
model.compile(optimizer="adam", loss=losses.MeanSquaredError(), metrics=['accuracy'])
model.fit(noizy_x_train, x_train, epochs=10, batch_size=64, validation_data=(noizy_x_test, x_test))
start = time.time()
predictions = model.predict(noizy_x_test)
print("Time Taken : %f" % (time.time() - start))
display(x_test, noizy_x_test, predictions)
```
